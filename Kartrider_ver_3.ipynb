```python
import pandas as pd
df = pd.read_csv('C:\\Users\\colacan\\Documents\\GitHub\\Kartrider_Meta_Analysis_Project\\kartrider_data.csv')
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
# 전처리 이전의 df
df
```

```python
# 결측치 비율 확인
df.isnull().sum()/113949
```

```python
# 필요없는 columns 드롭
df = df.drop(['matchResult','matchRetired','matchWin','playTime'],axis=1)
```

```python
# battle 데이터는 협력전이므로 제외
df= df[df['channelName'] != 'battle']
```

```python
# 결측치 제거
df = df.dropna(subset= ['kart', 'matchRank'])
```

```python
# 결측치 대체
values = {'pet':'non_pet', 'flyingPet':'non_flyingPet', 'partsEngine':'over_ver_engine', 'partsHandle':'over_ver_handle','partsWheel':'over_ver_wheel','partsKit':'over_ver_kit','matchTime': 999}
df = df.fillna(value=values)
```

```python
# matchTime second 단위로 통일
df['matchTime'] = df['matchTime'].astype(str)
df['matchTime'] = df.matchTime.str[0:3]
df['matchTime'] = df['matchTime'].astype(float)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# 인덱스 재정렬
df = df.reset_index()
df = df.drop('index',axis=1)
```

```python
# 전처리 이후의 df
df
```
